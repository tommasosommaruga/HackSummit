# Notebook 01 — Comtrade Trade Flow Exploration

Goal: identify countries where lithium exports exceed plausible supply (laundering hubs).

```python
import sys
sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
from src.model.mass_balance import multi_year_scan
```

```python
# Load seed data (CSV columns: reporter_code, year, flow_label, qty_kg)
trade_df = pd.read_csv('../data/raw/lithium_trade_flows.csv')
production_df = pd.read_csv('../data/raw/lithium_production.csv')

print(trade_df.head())
print(production_df.head())
```

```python
# Run mass-balance scan across 2019–2024
flagged = multi_year_scan(trade_df, production_df, years=list(range(2019, 2025)))
flagged
```

```python
# Visualise surplus by country and year
pivot = flagged.pivot_table(index='reporter_code', columns='year', values='surplus_kg', aggfunc='sum').fillna(0)
pivot.div(1e6).plot(kind='bar', figsize=(12, 5), title='Unexplained Lithium Surplus by Country (kt)')
plt.ylabel('Surplus (kt)')
plt.tight_layout()
plt.savefig('../data/outputs/laundering_surplus.png', dpi=150)
plt.show()
```
